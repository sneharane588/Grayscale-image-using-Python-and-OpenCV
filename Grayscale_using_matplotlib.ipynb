#### importing required python libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('mug2.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray', interpolation = 'bicubic')
plt.show()
```

```python
img1 = cv2.imread('mug.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img1,cmap='gray', interpolation = 'bicubic')
plt.plot([200,300,400],[100,200,300],'b',linewidth=5)
#plotting a line
#having width of 5
#having blue color
#points [x,x,x],[y,y,y]
plt.show()
```

```python
img2 = cv2.imread('watch.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img2,cmap='gray', interpolation = 'bicubic')
plt.plot([50,100],[80,100],'m',linewidth=4)
#plotting a line
#having width of 4
#having magenta color
#points [x,x],[y,y]
plt.show()
```

```python
img3 = cv2.imread('mug1.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img3,cmap='gray', interpolation = 'bicubic')
plt.xticks([]),plt.yticks([])
plt.plot([200,400,600],[500,600,700],'g',linewidth=7)
#        [x,x,x],[y,y,y]            green
plt.show()
```
